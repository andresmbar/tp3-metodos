# Precision

# Knn solo

```python
import csv

import pandas as pd
with open('../resultados/knn_solo_precision.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=(','), quotechar='|')
    numbers = next(spamreader)
    numbers.append('k')
    data = []
    k = 1 
    for row in spamreader:
        data_row = [float(i) for i in row]
        data_row.append(k)
        k += 500
        data.append(data_row)
df = pd.DataFrame(data, columns =numbers)
df =df.set_index('k')
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
image = sns.heatmap(df, annot=False,cmap="copper")
image.set(xlabel='Digito', ylabel='K',title= 'Precision de cada digito segun el k')
fig = image.get_figure()
fig.savefig('../resultados/knn_presicion', dpi =500)
```

## KNN + PCA

```python
import csv

import pandas as pd
with open('../resultados/knn_pca_precision.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=(','), quotechar='|')
    numbers = next(spamreader)
    data = []
    for row in spamreader:
        data_row = [float(i) for i in row]
        data.append(data_row)
df = pd.DataFrame(data, columns =numbers)
print(df)
```

### Alpha = 1 

```python
df_1 = df.loc[df['alpha'] == 1]
df_1 =df_1.set_index('k')
df_1.pop("alpha")

import seaborn as sns
import matplotlib.pyplot as plt
 
image = sns.heatmap(df_1, annot=False,cmap="RdBu")
image.set(xlabel='Digito', ylabel='K',title= 'Precision de cada digito segun el k para alpha = 1')
fig = image.get_figure()
fig.savefig('../resultados/knn_pca_presicion_1', dpi =500)
```
