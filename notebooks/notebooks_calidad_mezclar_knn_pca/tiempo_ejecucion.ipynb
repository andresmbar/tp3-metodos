# Knn + PCA

```python
import csv
import numpy as np
import pandas as pd
from statistics import mean

with open('../resultados/knn_pca_time_re_do.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=(','), quotechar='|')
    for row in spamreader:
       
        data_row = [float(i[1:-1]) for i in row]
        

data_divided_by_alpha = np.array_split(data_row, 7) 
data_mean_time = [ mean(i) for i in data_divided_by_alpha]
print(data_mean_time)
alphas = [1,5,9,13,17,21,25]
```

```python
import matplotlib.pyplot as plt


 
## Creamos Gráfica
plt.bar(alphas, data_mean_time)
 
## Legenda en el eje y
plt.ylabel('Tiempo promedio en segundos')
 
## Legenda en el eje x
plt.xlabel('Cantidad de componentes principales')
 
## Título de Gráfica
plt.title('Variacion del tiempo promedio para computar KNN + PCA segun los alphas')
 
## Mostramos Gráfica
plt.savefig('../resultados/knn_pca_tiempo', dpi =500,bbox_inches = "tight")
plt.show()
```
